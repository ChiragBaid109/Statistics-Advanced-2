Q1) What is hypothesis testing in statistics?

ans-Hypothesis testing in statistics is a method for making inferences about a population by examining data from a sample. It involves formulating a null hypothesis (a statement about the population) and an alternative hypothesis (a competing statement), then using sample data to determine whether there's enough evidence to reject the null hypothesis in favor of the alternative.

Q2) What is the null hypothesis, and how does it differ from the alternative hypothesis?

ans-In statistical hypothesis testing, the null hypothesis (H0) is a statement of no effect or no difference, while the alternative hypothesis (Ha or H1) proposes that there is a significant effect or difference. The null hypothesis serves as the default assumption that researchers try to disprove, while the alternative hypothesis represents the claim they are trying to prove.

Q3) Explain the significance level in hypothesis testing and its role in deciding the outcome of a test?

ans- In hypothesis testing, the significance level (denoted by Î±, or alpha) is a probability value that sets the threshold for determining whether to reject the null hypothesis

Q4) What are Type I and Type II errors? Give examples of each.

ans-In hypothesis testing, a Type I error (false positive) occurs when a true null hypothesis is incorrectly rejected, while a Type II error (false negative) occurs when a false null hypothesis is not rejected.

Q5) What is the difference between a Z-test and a T-test? Explain when to use each.

ans-The primary difference between a Z-test and a T-test lies in how they handle population standard deviation. Z-tests are used when the population standard deviation is known, while T-tests are used when it is unknown and must be estimated from the sample.

```python
"""7 Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
"""
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu = 50
sigma = 0.5
n = len(sample_data)
sample_mean = np.mean(sample_data)
z = (sample_mean - mu) / (sigma / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Z-statistic = {z:.4f}")
print(f"P-value = {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")
```

```python
#8 Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1. Simulate data
np.random.seed(0)  # for reproducibility
mean = 100         # true population mean
std_dev = 15       # population standard deviation
n = 50             # sample size

data = np.random.normal(loc=mean, scale=std_dev, size=n)

# 2. Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
standard_error = sample_std / np.sqrt(n)

# 3. Calculate the 95% confidence interval (using t-distribution)
confidence = 0.95
df = n - 1
t_critical = stats.t.ppf((1 + confidence) / 2, df)
margin_of_error = t_critical * standard_error

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# 4. Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# 5. Plot the data
plt.figure(figsize=(10, 5))
plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle=':', label=f'95% CI Lower = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle=':', label=f'95% CI Upper = {ci_upper:.2f}')
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
"""9 : Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
"""
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot a histogram.
    """
    mean = np.mean(data)
    std = np.std(data)

    # Avoid division by zero
    if std == 0:
        raise ValueError("Standard deviation is zero. Z-scores cannot be computed.")

    z_scores = (data - mean) / std

    # Plot histogram
    plt.figure(figsize=(10, 5))
    plt.hist(z_scores, bins=10, color='lightcoral', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score (Standard Deviations from Mean)')
    plt.ylabel('Frequency')
    plt.axvline(0, color='blue', linestyle='--', label='Mean (Z=0)')
    plt.axvline(1, color='green', linestyle='--', label='1 Std Dev')
    plt.axvline(-1, color='green', linestyle='--')
    plt.axvline(2, color='orange', linestyle='--', label='2 Std Dev')
    plt.axvline(-2, color='orange', linestyle='--')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return z_scores

# Example usage
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=100)
z_scores = calculate_z_scores(data)

# Optional: print a few Z-scores
print("First 5 Z-scores:", np.round(z_scores[:5], 2))
```
